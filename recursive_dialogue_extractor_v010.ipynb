```python
# s: the start-point of the window; w: the width of the window; the end-point of the window will be s+w
def dialogue(text, characters, dialogue_=[]):
    s, w = 1, 2  # starting width of the window. we shouldn't start from zero!
    while s < len(text):  # until we have some text to search inside
        if text[s:s+w] in characters:  # if we found a character in the sliding window
            dialogue_.append(text[:s])  # append the text before the new-found character to the dialogue_
            dialogue(text[s:], characters, dialogue_=dialogue_)  # call the function again on the remaining text, and append to the dialogue_
        if s + w >= len(text):  # if end-point reached the end of the text, start from the next character and reset the width of the window to 2
            s += 1
            w = 2
        else:  # widen the window by one
            w += 1
    dialogue_.append(text)
    return(dialogue_)
```
